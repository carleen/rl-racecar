```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
ls = pd.read_csv('l_sarsa.csv')
lq = pd.read_csv('results-q_l.csv')
lq.head()
```

```python
'''
Test statistics for the experiments:

Check the distribution of the number of timesteps for each of the algorithms
'''
def generateGraphs(ls, lq, experiment_name):
    width=800
    height=650
    labels = {'num_timesteps': 'Number Timesteps Required', 'probability density': 'Percentage of Games'}

    fig_ls = px.histogram(ls, x="num_timesteps",
                      title="Distribution of number of timsteps, SARSA Model",
                         width = width,
                         height=height,
                         histnorm="probability density",
                         labels=labels)
    fig_lq = px.histogram(ls, x="num_timesteps",
                      title="Distribution of number of timsteps, Q-Learning Model",
                          width = width,
                         height=height,
                         histnorm="probability density",
                         labels=labels)

    combined = pd.DataFrame({'SARSA': ls.num_timesteps,
                            'Q-Learning': lq.num_iterations})

    fig = px.ecdf(combined, x=['SARSA', 'Q-Learning'],
                  labels={'variable': 'Number Timesteps Required',
                         'value': 'Number of Timesteps',
                         'probability': 'Probability'}, 
                  width=width, height=height, 
                  range_x=[0,250],
                 marginal="histogram")
    fig.write_image(f'{experiment_name}.png')
    
    return fig
```

```python
generateGraphs(ls, lq, 'L_track')
```

```python
ls = pd.read_csv('o_sarsa.csv')
lq = pd.read_csv('results-q_o.csv')
f = generateGraphs(ls, lq, 'O_track')
```
